```python
from random import randint
import time
import math
```

```python
n =  2

p = {}

p[0] = 0
p[1] = 1

#randomly creating length vs price mapping
for i in range(2,10000):
    p[i] = randint(i,3*i)
```

```python
def naive_rod_cutting(n,p):
    
    #print(n)
    if n == 0:
        return 0

    
    q = -1
    
    for i in range(1,n+1):
        #print(i,n-i)
        q = max(q,p[i] + naive_rod_cutting(n-i,p))
        #print(q)
    return q
```

```python
time_naive = []
for i  in range(1,28):
    print(i)
    start = time.time()
    naive_rod_cutting(i,p)
    end = time.time()
    time_naive.append(end-start)
```

## The power of Dynamic Programming

```python
#Top Down Approach


def memoized_cut_rod(n,p):
    
    r = {}
    for i in range(n+1):
        r[i] = -1
        
    return memoized_cut_rod_aux(p,n,r)

def memoized_cut_rod_aux(p,n,r):
    
    if r[n] >= 0:
        return r[n]
    
    if n == 0:
        
        q = 0
    else:
        q = -1
        for i in range(1,n+1):
            q = max(q,p[i] + memoized_cut_rod_aux(p,n-i,r))
        
    r[n] = q
    
    return q
```

```python
time_dp = []
for i  in range(1,900):
    print(i)
    start = time.time()
    memoized_cut_rod(i,p)
    end = time.time()
    time_dp.append(end-start)
```

```python
#Bottom up approach

def bottom_up_cut_rod(n,p):
    
   
    
    r = {}
    r[0] = 0
    
    
    for j in range(1,n+1):

        q = -1
        for i in range(1,j+1):

            
            q = max(q,p[i] + r[j-i])
            
        r[j] = q

        
    return r[n]
```
